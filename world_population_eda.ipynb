```python
## 2. Data Cleaning & Transformation
#Before analyzing the data, we need to ensure it is clean. 
#We will handle any missing values, enforce the correct data types, engineer a new growth rate feature,
#reshape the data to make time-series plotting easier later on.
```

```python
# CELL 8: Handle missing values and fix data types
# Based on our initial inspection, the data contains 0 null values 
# and the population columns are already integers (int64). No imputation needed!

print(f"Total missing values in dataset: {df.isnull().sum().sum()}")
```

```python
# CELL 9: Create a new derived column for Total Growth from 1970 to 2022
# Formula: ((New - Old) / Old) * 100

df["Total Growth 1970-2022 (%)"] = ((df["2022 Population"] - df["1970 Population"]) / df["1970 Population"]) * 100

# Let's check the result by looking at just a few relevant columns
df[["Country/Territory", "1970 Population", "2022 Population", "Total Growth 1970-2022 (%)"]].head()
```

```python
# CELL 10: Reshape the dataframe using pd.melt()
# We will keep the descriptive columns as identifiers and melt the population year columns.

# 1. Define the columns that identify the country
id_cols = ['Rank', 'CCA3', 'Country/Territory', 'Capital', 'Continent', 'Area (km²)', 'Density (per km²)', 'Growth Rate', 'World Population Percentage', 'Total Growth 1970-2022 (%)']

# 2. Define the columns we want to melt (the years)
year_cols = ['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population']

# 3. Perform the melt
df_melted = pd.melt(df, 
                    id_vars=id_cols, 
                    value_vars=year_cols, 
                    var_name='Year', 
                    value_name='Population')

# 4. Clean up the 'Year' column: remove the word " Population" and convert to integer for proper time-series plotting
df_melted['Year'] = df_melted['Year'].str.replace(' Population', '').astype(int)

# Check the new long-format dataframe
df_melted.head()
```

```python
# CELL 11: Top 10 most populous countries in 2022
top_10_2022 = df.nlargest(10, "2022 Population")

plt.figure(figsize=(10, 6))
# Using a clean color palette and plotting horizontal bars
ax = sns.barplot(x="2022 Population", y="Country/Territory", data=top_10_2022, palette="viridis")

# Add value labels to the end of each bar for clarity
for i in ax.containers:
    ax.bar_label(i, fmt='%.0f', padding=3)

plt.title("Top 10 Most Populous Countries (2022)", fontsize=14, fontweight='bold')
plt.xlabel("Population (in Billions)")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
# CELL 12: World population growth over time
# Group the melted dataframe by Year and sum the global population
world_pop_trend = df_melted.groupby('Year')['Population'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Population', data=world_pop_trend, marker='o', color='#2b8cbe', linewidth=2.5)

# Annotating a key milestone
plt.annotate('Approaching 8 Billion', 
             xy=(2022, 7970000000), 
             xytext=(2000, 7500000000),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8), 
             fontsize=11, fontweight='bold')

plt.title("Global Population Growth (1970 - 2022)", fontsize=14, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Total Global Population")
plt.show()
```

```python
# CELL 13: Continent-level comparison over the decades
# Group by both Continent and Year
continent_pop = df_melted.groupby(['Continent', 'Year'])['Population'].sum().reset_index()

plt.figure(figsize=(14, 7))
sns.barplot(x='Year', y='Population', hue='Continent', data=continent_pop, palette='Set2')

plt.title("Population Growth by Continent (1970 - 2022)", fontsize=14, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Total Population")
plt.legend(title='Continent', loc='upper left')
plt.show()
```

```python
# CELL 14: Correlation heatmap of numeric variables
plt.figure(figsize=(10, 8))

# Isolate only the numeric columns so the correlation matrix doesn't throw an error on text columns
numeric_df = df.select_dtypes(include=[np.number])

# Plot the heatmap
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of Demographic Variables", fontsize=14, fontweight='bold')
plt.show()
```

```python
# CELL 15: Fastest growing vs. declining countries (1970-2022)
# Grab the top 10 and bottom 10 based on our custom feature
top_growth = df.nlargest(10, 'Total Growth 1970-2022 (%)')
bottom_growth = df.nsmallest(10, 'Total Growth 1970-2022 (%)')

# Combine them into one dataframe
extreme_growth = pd.concat([top_growth, bottom_growth]).sort_values(by='Total Growth 1970-2022 (%)', ascending=False)

plt.figure(figsize=(12, 8))

# Create a custom color list: Green for positive growth, Red for negative/low growth
colors = ['#2ca02c' if x > 0 else '#d62728' for x in extreme_growth['Total Growth 1970-2022 (%)']]

sns.barplot(x='Total Growth 1970-2022 (%)', y='Country/Territory', data=extreme_growth, palette=colors)

plt.title("Fastest Growing vs. Most Declining Populations (1970 - 2022)", fontsize=14, fontweight='bold')
plt.xlabel("Total Growth Rate (%)")
plt.ylabel("")
plt.axvline(0, color='black', linewidth=1.5) # Distinct line at 0%
plt.tight_layout()
plt.show()
```

```python
## ## Conclusions & Key Insights

# After loading, cleaning, and visualizing the World Population dataset (1970–2022), several critical demographic trends emerged:

# **Unprecedented Global Growth:** The global population has experienced an incredibly steep, near-linear ascent over the last 52 years, effectively doubling from roughly 3.7 billion in 1970 to nearly 8 billion in 2022.
# **Continental Asymmetry:** Asia remains the absolute demographic giant, anchoring over 59% of the world's population. However, while Asia's absolute numbers are massive, Africa exhibits the steepest relative acceleration curve across the decades, contrasted sharply against Europe’s stagnant growth.
# **The Scale of Dominance:** The top two most populous nations (China and India) both exceed 1.4 billion residents, completely altering the scale of global demographics and visually dwarfing the rest of the top 10 countries combined.
# **Explosive Outliers vs. Regional Decline:** Feature engineering revealed extreme regional variance. Geopolitical and economic hubs in the Middle East (specifically the United Arab Emirates and Qatar) saw explosive outlier growth exceeding 1,000% since 1970 due to immigration and economic expansion. Conversely, multiple Eastern European nations are experiencing clear, measurable population declines.
# **Growth Deceleration:** The correlation analysis confirms a negative relationship between total population size and current growth rates, signaling that while the population is still growing, the *velocity* of that growth is beginning to cool down globally.
```
